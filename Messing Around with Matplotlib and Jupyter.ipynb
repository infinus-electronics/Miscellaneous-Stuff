````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
%matplotlib inline


fig, ax = plt.subplots()
x = np.linspace(-np.pi, np.pi, 201) #linspace generates an nd array of a specified number of samples within the 2 boundaries

ax.plot(x, np.sin(x))
ax.set_title("Sine Graph Tester")
ax.set_xlabel("Angle[rad]")
ax.set_ylabel("Magnitude")
ax.axis("tight")
````

````python
#fig2 = plt.figure()

y = np.linspace(-np.pi, np.pi, 201)
plt.plot(y, np.cos(y))
plt.title("Cosine Graph Tester")
plt.xlabel("Angle[rad]")
plt.ylabel("Magnitude")
plt.axis("tight")
````

````python
plt.plot(x, np.sin(x), 'r.') # plot dots
plt.show()
plt.xlim([-0.5, 0.5]) # set x axis limits
plt.plot(x, np.sin(x), 'r.') # plot dots
plt.title('Zoomed in')
plt.show()
plt.plot(x, np.cos(x), 'g-') # plot bars (too small to see)
plt.show()
plt.plot(x, np.sin(x), 'b')
plt.show()
````

````python
figure, axis = plt.subplots(2,2) # 2x2 subplot thing
figure.subplots_adjust(bottom = 0.1, right = 2.0, top = 3)
axis[0][0].plot(x, np.sin(x))
axis[0][0].set_xlabel('x')
axis[0][0].set_title('Sine Graph')
axis[0][0].set_ylabel('sin(x)')
axis[0][1].plot(x, np.cos(x))
axis[1][0].plot(x, np.tan(x))
axis[1][0].set_ylim([-1,1])
axis[1][0].set_title('Tangent Graph')
axis[1][1].plot(x, np.sin(x), 'r')
axis[1][1].plot(x, np.cos(x), 'g')
````

````python
figure, axis = plt.subplots(2,2) # 2x2 subplot thing
figure.subplots_adjust(bottom = 0.1, right = 2.0, top = 3)
axis[0][0].plot(x, np.sin(x))
axis[0][0].set_xlabel('x')
axis[0][0].set_title('Sine Graph')
axis[0][0].set_ylabel('sin(x)')
axis[0][1].plot(x, np.cos(x))
axis[1][0].plot(x, np.tan(x))
axis[1][0].set_title('Tangent Graph')
axis[1][0].set_ylim([-1,1])
axis[1][1].plot(x, np.sin(x), 'r')
axis[1][1].plot(x, np.cos(x), 'g')

plt.tight_layout()
````

````python
x = np.linspace(-np.pi, np.pi, 51)
plt.plot(x, np.sin(x), 'r.', label = "sin(x)")
plt.plot(x, np.cos(x), label = "cos(x)")
plt.legend(loc = 'lower right')
````
